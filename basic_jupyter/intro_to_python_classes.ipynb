```python
import math

class Point(object):

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
       return "Point(%d, %d)" % (self.x, self.y)

    def distance_from_origin (self):
        return math.sqrt(self.x**2 + self.y**2)

    def distance(self, p2=None):
        if p2 is None:
            p2 = Point(0,0)
        return math.sqrt((p2.x - self.x)**2 + (p2.y - self.y)**2)
```

```python
p1 = Point(3,4)
p2 = Point(1,1)
```
